```python
from pathlib import Path

RAW_ROOT = Path("../data/raw/NEST 2.0 Data files_Anonymized")
print(RAW_ROOT)
print("Exists:", RAW_ROOT.exists())
print("Is directory:", RAW_ROOT.is_dir())
```

```python
xlsx_files = list(RAW_ROOT.rglob("*.xlsx"))
len(xlsx_files)
```

```python
# File Type Classifier
def classify_file(path: Path):
    name = path.name.lower()
    if "cpid_edc_metrics" in name:
        return "cpid"
    if "visit projection" in name:
        return "visit_projection"
    if "missing_lab" in name:
        return "lab_missing"
    if "sae" in name:
        return "sae"
    if "missing_pages" in name:
        return "missing_pages"
    if "compiled_edrr" in name:
        return "edrr"
    if "meddra" in name:
        return "meddra"
    if "who" in name:
        return "whodrug"
    return "other"
```

```python
# Inventory Table
inventory = pd.DataFrame({
    "file_path": [str(f) for f in xlsx_files],
    "file_name": [f.name for f in xlsx_files],
    "file_type": [classify_file(f) for f in xlsx_files]
})

print(inventory["file_type"].value_counts())
print("Total = " + str(sum(inventory["file_type"].value_counts())))
```

```python
# Save inventory to CSV
inventory.to_csv("../data/intermediate/file_inventory.csv", index=False)
```
